```python
import pandas as pd
import numpy as np

df = pd.read_csv('SLR.csv', encoding='utf-8')
```

```python
df = df.drop(['state'], axis=1)
```

```python
df
```

```python
for k in range(len(df)):
    print("{", end="")
    for index, value in df.loc[k].items():
        if str(value) != 'nan':
            if type(value) == str:
                print("'"+index+"'"+":", "'"+value.upper()+"'", end=", ")
            else :
                print("'"+index+"'"+":", int(value), end=", ")
    print("},")
```
